```python
'''
This function runs the router api on all of the coordinate pairs we create above in 
the list xytups and writes the travel times to a csv
'''
import numpy as np
def travel_time(index,lenmatrix):
    #create a 1 x length of xytups matrix
    mat= np.zeros((1,len(xytups)))
    mat1=np.zeros((1,len(xytups)))
    # for the length of the matrix
    for j in range(0,lenmatrix):
        #origin lat
        ox= xytups[index][0]
        #origin long
        oy= xytups[index][1]
        #destination lat
        dx= xytups[j][0]
        #destination long
        dy= xytups[j][1]
        #coordinates in string form
        cords= [str(oy),str(ox),str(dy),str(dx)]
        #run the api on these coords
        a= our_api(cords)
        #put the travel time in matrix
        mat[0][j] =a
        #2nd point
        ox1= xytups[index+1][0]
        oy1= xytups[index+1][1]
        dx1= xytups[j][0]
        dy1= xytups[j][1]
        cords1= [str(oy1),str(ox1),str(dy1),str(dx1)]
        a1= our_api(cords1)
        mat1[0][j] =a1
    #puts our list in a numpy array
    fin_list= np.array(mat)
    fin_list=np.append(mat,mat1,axis=0)
    #save numpy array as csv file
    np.savetxt("all_points_tst" + str(index) + ".csv", fin_list, delimiter =",")
```

```python
#helpfule function for running our API
#cord_list = [strox, stroy, strdx, strdy]
#aka format for input is stroy, stroc, strdy, strdx
def our_api(cord_list): 
    url1 = 'http://127.0.0.1:5000/route/v1/driving/' + cord_list[1] + ',' + cord_list[0] + ';' + cord_list[3] + ',' + cord_list[2]+'?steps=false'
    
    try: 
        with urllib.request.urlopen(url1) as url:
            data = json.loads(url.read().decode())
        ortime= (data.get('routes')[0].get('duration'))
    except Exception:
        try:
            match_url = 'http://127.0.0.1:5000/nearest/v1/driving/' + cord_list[3] + ',' + cord_list[2]
            with urllib.request.urlopen(match_url) as url2:
                data2 = json.loads(url2.read().decode())
            new_cords = (data2.get('waypoints')[0].get('location'))
            str0 = str(new_cords[0])
            str1 = str(new_cords[1])
            url3 = 'http://127.0.0.1:5000/route/v1/driving/' + cord_list[1] + ',' + cord_list[0] + ';' + str0 + ',' + str1 +'?steps=false'
            with urllib.request.urlopen(url3) as url4:
                data3 = json.loads(url4.read().decode())
            ortime = (data3.get('routes')[0].get('duration'))
        except Exception:
            ortime = 0.111
   
    return ortime 
```

```python
'''
This function allows you to choose the amount of threads you would like to run the 
router on
'''
#imports are for threading and for timing the process
import os
import threading
import time
def main2(num,threads):
    #get start time
    t1=time.time()
    #incrementer
    a=0
    #need an empty list to append the threads to
    thr=[]
    #loop until we have numxnum travel times
    while a<num:
        #loop until the the amount of threads have been created and started
        for i in range(0,threads):
            #try doing this
            try:
                #create thread to run travel_time
                t=threading.Thread(target=travel_time, args=(a+i,num))
                #append the thread to a list
                thr.append(t)
                #start the thread
                t.start()
            except Exception as e:
                print(e)
                continue
        #increment the incrementer by the amount of threads
        a+=threads
    #get the time it ended     
    t2=time.time()-t1
    print(t2)
    print(threads)
```
